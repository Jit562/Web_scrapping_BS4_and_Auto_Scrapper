## What is Web Scraping



Web scraping using Python refers to the process of automatically extracting data from websites. It involves writing code to navigate web pages, retrieve specific information, and store or manipulate that data for various purposes. Web scraping is commonly used for tasks such as data collection, data analysis, content extraction, and more.

Here's a basic overview of how web scraping using Python works:

1. HTTP Requests: Python libraries like requests or urllib are used to send HTTP requests to a target website. These requests simulate the behavior of a web browser and fetch the HTML content of web pages.

2. Parsing HTML: Once the HTML content is retrieved, Python libraries like Beautiful Soup or lxml are often employed to parse the HTML and navigate the page's structure. These libraries provide methods to access specific HTML elements and their attributes.

3. Extracting Data: Web scraping involves identifying the elements on a web page (e.g., headings, paragraphs, tables) that contain the data you want to extract. You can use the parsed HTML and CSS selectors to locate and extract relevant information.

4. Data Cleaning: The extracted data may require cleaning and formatting to remove unwanted characters, spaces, or HTML tags. Python provides powerful string manipulation functions for this purpose.

5. Data Storage: You can store the extracted data in various formats, such as CSV, JSON, databases, or even in-memory data structures like lists or dictionaries, depending on your needs.

6. Automation: To scrape multiple pages or perform repeated scraping tasks, you can automate the process by using loops, conditionals, and functions in your Python code.

7. Handling Dynamic Content: Some websites use JavaScript to load content dynamically after the initial page load. In such cases, you may need to use tools like Selenium or Puppeteer, which can interact with the webpage, execute JavaScript, and retrieve the dynamic content.

It's important to note that web scraping should be done responsibly and ethically. Always check a website's robots.txt file to see if scraping is allowed or restricted. Respect the website's terms of service and avoid causing excessive traffic or disrupting its functionality. Additionally, scraping large amounts of data from a website without permission may be illegal in some jurisdictions.

## Beautiful Soup Documentation

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
webpage = requests.get('https://www.webscraper.io/test-sites/e-commerce/allinone/computers/tablets').text
```

```python
soup = BeautifulSoup(webpage, "lxml")
```

```python
#print(soup.prettify())
names = soup.find_all("a", class_="title")
```

```python
name = []
for i in names:
  name.append(i.text)
```

```python
name
```

```python
prices = soup.find_all("h4", class_="price")
```

```python
price = []

for i in prices:
  price.append(i.text)
```

```python
price
```

```python
desc = soup.find_all("p", class_="description")

description = []

for i in desc:
  description.append(i.text)

description
```

```python
reviews = soup.find_all("p", class_="review-count")

review = []

for i in reviews:
  review.append(i.text)

review
```

## Create the data frame

```python
df = pd.DataFrame({'name':name, 'price':price, 'description':description, 'review':review})
```

```python
df.head(4)
```

## Ipl Auction data extract

```python
ipl = requests.get('https://www.iplt20.com/auction').text
```

```python
soupp = BeautifulSoup(ipl, "lxml")
```

```python
#print(soupp)

team_name = soupp.find_all("div", class_="agv-team-name")

team_names = []
for i in team_name:
  team_names.append(i.text)


team_names
```

```python
fund_reaminigs = soupp.find_all("div", class_="avg-fund-remaining")


fund_remaining = []

for i in fund_reaminigs:
  fund_remaining.append(i.text[17:-2])

fund_remaining
```

```python
overseas = soupp.find_all("li", class_="m-0")

over_sease = []
over_sease_player = []

for i in overseas:
  over_sease.append(i.text[-2:-1])


for x in over_sease:
  if '8' in x:
    over_sease_player.append(x)

over_sease_player
```

```python
totalplayer = soupp.find_all("li", class_="m-0 px-1")


total_player = []


for i in totalplayer:
  total_player.append(i.text[-3:-1])

total_player
```

## Create dataframe for ipl data auction data

```python
ipn_data = pd.DataFrame({'team_names':team_names, 'fund_remaining':fund_remaining , 'over_sease_player':over_sease_player, 'total_player':total_player})
```
